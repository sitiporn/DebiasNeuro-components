```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import yaml
```

```python
with open('../../pickles/losses/reweight.pickle', 'rb') as handle:
    losses = pickle.load(handle)

with open("../config.yaml", "r") as yamlfile:
    config = yaml.load(yamlfile, Loader=yaml.FullLoader)
```

```python
title = f"Minize loss on vadiation set MNLI-Matched set using {config['dev-name']}"

plt.title(title)
plt.plot(losses, label='loss')
plt.xlabel("n iteration")
plt.ylabel("Loss")
plt.show()
```
